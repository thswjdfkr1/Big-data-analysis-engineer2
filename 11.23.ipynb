```python
import pandas as pd
```

```python
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_test.csv")
```

```python
train = pd.merge(x_train, y_train, on='StudentID')
target = x_test.copy()
```

```python
print(train.shape)
print()
print(train.info())
print()
print(train.describe())
print()
print(train.isnull().sum())
```

```python
print(target.shape)
print()
print(target.info())
print()
print(target.describe())
print()
print(target.isnull().sum())
```

```python
x = train.drop(['StudentID','G3'], axis=1)
y = train['G3']
```

```python
obj_col = x.select_dtypes(include='object').columns
```

```python
for col in obj_col:
  print(col)
  print(x[col].unique())
```

```python
for col in obj_col:
  print(col)
  print(target[col].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_col = ['Mjob','Fjob']
```

```python
for col in le_col:
  le = LabelEncoder()
  x[col] = le.fit_transform(x[col])
  target[col] = le.transform(target[col])
```

```python
print(x.head())
print(obj_col)
```

```python
ohe_col = obj_col.drop(le_col)
```

```python
x = pd.get_dummies(x, columns = obj_col, dtype=int)
target = pd.get_dummies(target, columns = obj_col, dtype=int)
```

```python
# from sklearn.preprocessing import MinMaxScaler

# mms = MinMaxScaler()
# x[mms_col] = mms.fit_transform(x[mms_col])
# target[mms_col] = mms.transform(target[mms_col])
```

```python
from sklearn.model_selection import train_test_split

rs = 810
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=rs)

print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=rs)
rf.fit(x_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

param = {'n_estimators':[5,10,15,20],
         'max_depth':[5,10,15,20]}

gscv = GridSearchCV(estimator= rf,
                    param_grid=param,
                    cv=5)
gscv.fit(x_train, y_train)

model = gscv.best_estimator_
print(gscv.best_params_)
```

```python
pred_val = model.predict(x_val)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

print('mse', mean_squared_error(y_val, pred_val))
print('mse', mean_squared_error(y_val, pred_val, squared=False))
print('mse', np.sqrt(mean_squared_error(y_val, pred_val)))
print('mae', mean_absolute_error(y_val, pred_val))
```

```python
pred = model.predict(target.drop(['StudentID'], axis=1))
```

```python
result = pd.DataFrame({'ID':target['StudentID'],
                      'pred':pred})

result.to_csv('result.csv', index=False)
print(result)
```

```python
end = channel[channel.ct.dt.strftime('%Y-%m-%d %H') =='2021-11-01 15']
start = channel[channel.ct.dt.strftime('%Y-%m-%d %H') =='2021-10-03 03']

end_df = end[['channelname','subcnt']].reset_index(drop=True)
end_df.columns = ['channelname','subcnt_end']
start_df = start[['channelname','subcnt']].reset_index(drop=True)
start_df.columns = ['channelname','subcnt_start']

target = pd.merge(end_df,start_df, on='channelname')
target['del'] = target['subcnt_end'] - target['subcnt_start']
print(target[['channelname','del']])
```

```python
import pandas as pd

df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
ans = df.groupby('Country').sum().sort_values('Goals',ascending=False)
result = ans[['Goals']].head(5)
print(result)
```

```python
ans = df.groupby('Country').size().sort_values(ascending=False)
print(ans.head())
```

```python
df['nn'] = df.Years.str.split('-')

def cc(x):
  for i in x:
    if len(str(i)) != 4:
      return False

  return True

df['cc'] = df['nn'].apply(cc)
print(len(df[df['cc'] == False]))
```

```python
df2 = df[df['cc'] == True].reset_index()
print(df2.shape[0])
```

```python
df2['lencup'] = df2['nn'].str.len()
ans = len(df2[df2['lencup'] == 4])
print(ans)
```
